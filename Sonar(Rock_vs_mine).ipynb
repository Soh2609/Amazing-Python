```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
#load dataset to pandas dataframe
sonar_data = pd.read_csv('/content/sonar data.csv',header=None)
```

```python
sonar_data.head() #determine first 5 rows
```

```python
sonar_data.shape #determine number of rows and column
```

```python
sonar_data.describe() #determine the statistical measures
```

```python
sonar_data[60].value_counts() #counting rock and mine examples
```

```python
sonar_data.groupby(60).mean()
```

```python
#separating data and labels
X =sonar_data.drop(columns=60,axis=1)
Y =sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

```python
# training and testing
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train,Y_train)
```

```python
#model training
model=LogisticRegression()
```

```python
#training model with data
model.fit(X_train,Y_train)
```

```python
#evaluation # accuracy for training
X_train_prediction=model.predict(X_train)
accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("accuracy is", accuracy)
```

```python
X_test_prediction=model.predict(X_test)  # accuracy for testing
accuracy_testing=accuracy_score(X_test_prediction,Y_test)
```

```python
print("accuracy is", accuracy_testing)
```
